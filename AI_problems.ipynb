# 8-Puzzle problem

```python
class Puzzle:
  def __init__(self,my_list,goal_list):
    self.current_state=my_list
    self.goal_state=goal_list

  def move_empty(self,empty_pos):
    if empty_pos==0:
      return [1,3]
    elif empty_pos==1:
      return [0,2,4]
    elif empty_pos==2:
      return [1,5]
    elif empty_pos==3:
      return [0,4,6]
    elif empty_pos==4:
      return [1,3,5,7]
    elif empty_pos==5:
      return [2,4,8]
    elif empty_pos==6:
      return [3,7]
    elif empty_pos==7:
      return [4,6,8]
    elif empty_pos==8:
      return [5,7]

  def print_state(self):
    for i in range(0,8,3):
      print(self.current_state[i],"|",self.current_state[i+1],"|",self.current_state[i+2])
      if(i<6):
        print("__ __ __\n")


  def check_goal(self):
    if(self.current_state==self.goal_state):
      return True
    else:
      return False

  def findIdxofEmpty(self):
    if(' ' in self.current_state):
      index=self.current_state.index(' ')
      return index
    else:
      return -1

  def swap(self,idx1,idx2):
    temp=self.current_state[idx1]
    self.current_state[idx1]=self.current_state[idx2]
    self.current_state[idx2]=temp

  def play(self):
    while(self.check_goal()==False):
      possible_move=self.move_empty(self.findIdxofEmpty())
      move=int(input("Enter the move: "))
      while(move not in possible_move):
        print("Invalid move!")
        move=int(input("Enter the move: "))
      print("move ' ' to position : ",move)
      self.swap(self.findIdxofEmpty(),move)
      self.print_state()
    print("goal reached!")

my_list=[1,2,3,7,8,4,' ',6,5]
goal_list=[1,2,3,8,' ',4,7,6,5]
obj=Puzzle(my_list,goal_list)
obj.print_state()
obj.play()


```

## Water-Jug Problem

```python
class WaterJug:
  def __init__(self,capacity,initial,goal):
    self.jug1=capacity[0]
    self.jug2=capacity[1]
    self.current_state=initial
    self.goal_state=goal

  def perform_move(self,move):
      x, y = self.current_state
      if move == 1:    # Fill jug1
          x = self.jug1
      elif move == 2:  # Fill jug2
          y = self.jug2
      elif move == 3:  # Empty jug1
          x = 0
      elif move == 4:  # Empty jug2
          y = 0
      elif move == 5:  # Pour water from jug1 to jug2
          transfer = min(x, self.jug2 - y)
          x -= transfer
          y += transfer
      elif move == 6:  # Pour water from jug2 to jug1
          transfer = min(y, self.jug1 - x)
          y -= transfer
          x += transfer

      self.current_state = [x, y]


  def print_state(self):
    print(f"[{self.current_state[0]},{self.current_state[1]}]")

  def check_goal(self):
    return self.current_state==self.goal_state

  def play(self):
        print("Initial State:")
        self.print_state()
        print("Goal State:")
        print(f"[{self.goal_state[0]}, {self.goal_state[1]}]")
        print("Moves:")
        print("1: Fill jug1 \n2: Fill jug2 \n3: Empty jug1 \n4: Empty jug2 \n5: Pour water from jug1 to jug2 \n6: Pour water from jug2 to jug1")
        while not self.check_goal():
            print("\nCurrent State:")
            self.print_state()
            move = int(input("Enter your move (1-6): "))
            while move not in [1, 2, 3, 4, 5, 6]:
                print("Invalid move! Try again.")
                move = int(input("Enter your move (1-6): "))
            self.perform_move(move)
        print("\nGoal reached!")
        self.print_state()

capacity_list=[3,5]
initial_list=[1,2]
goal_list=[3,4]
obj=WaterJug(capacity_list,initial_list,goal_list)
obj.play()
```

# Travelling Salesman Problem

```python
from itertools import permutations

class TravellingSalesman:
    def __init__(self, cities, distances):
        self.cities = cities
        self.distances = distances
        self.num_cities = len(cities)

    def calculate_distance(self, path):
        total_distance = 0
        for i in range(len(path) - 1):
            from_city = path[i]
            to_city = path[i + 1]
            total_distance += self.distances[from_city][to_city]
        # Add the distance to return to the starting city
        total_distance += self.distances[path[-1]][path[0]]
        return total_distance

    def solve(self):
        city_indices = list(range(self.num_cities))
        all_paths = permutations(city_indices)
        min_distance = float('inf')
        best_path = None
        for path in all_paths:
            distance = self.calculate_distance(path)
            if distance < min_distance:
                min_distance = distance
                best_path = path
        return best_path, min_distance

    def print_solution(self, best_path, min_distance):
        city_names = [self.cities[i] for i in best_path]
        city_names.append(self.cities[best_path[0]])
        print("Shortest Path:", " -> ".join(city_names))
        print("Total Distance:", min_distance)

cities = ["1","2","3","4"]
distances = [
    [0, 10, 15, 20],  # Distances from 1
    [10, 0, 35, 25],  # Distances from 2
    [15, 35, 0, 30],  # Distances from 3
    [20, 25, 30, 0]   # Distances from 4
]

obj = TravellingSalesman(cities, distances)
best_path,min_distance = obj.solve()
obj.print_solution(best_path, min_distance)
```

# AI Water Jug problem

```python
from copy import deepcopy

class WaterJug:
    def __init__(self,capacity,initial,goal):
        self.jug1 = capacity[0]
        self.jug2 = capacity[1]
        self.current_state = tuple(initial)
        self.goal_state = tuple(goal)

    def get_next_states(self,state):
        x,y = state
        possible_states = []
        possible_states.append((self.jug1, y))
        possible_states.append((x, self.jug2))
        possible_states.append((0, y))
        possible_states.append((x, 0))
        transfer = min(x, self.jug2 - y)
        possible_states.append((x - transfer, y + transfer))
        transfer = min(y, self.jug1 - x)
        possible_states.append((x + transfer, y - transfer))
        return possible_states

    def print_state(self, state):
        print(f"[{state[0]}, {state[1]}]")

    def dfs(self):
        open_list = [(self.current_state, None)]
        closed_list = set()
        parents = {}
        while open_list:
            state, parent = open_list.pop()
            if state in closed_list:
                continue
            closed_list.add(state)
            parents[state] = parent
            if state == self.goal_state:
                print("\nGoal reached!")
                path = self.reconstruct_path(parents, state)
                for node in path:
                    self.print_state(node)
                return
            for next_state in self.get_next_states(state):
                if next_state not in closed_list:
                    open_list.append((next_state, state))
        print("No solution found.")

    def reconstruct_path(self, parents, goal_state):
        path = []
        current = goal_state
        while current is not None:
            path.append(current)
            current = parents[current]
        path.reverse()
        return path

capacity_list = [3,5]
initial_list = [1,2]
goal_list = [3,4]
obj = WaterJug(capacity_list, initial_list, goal_list)
obj.dfs()
```
